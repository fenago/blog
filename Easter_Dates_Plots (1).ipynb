# Plots

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=[12, 8], facecolor='limegreen')
plt.bar(data['DateRaw'].value_counts().reindex(reind).index, data['DateRaw'].value_counts().reindex(reind).values,color='deeppink')
plt.axhline(y=72,linewidth=1, color='blue', label='72')
plt.xticks(rotation=45)
plt.title('Easter: 1583 to 4099')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[12, 8], facecolor='limegreen')
bars = plt.bar(data['DateRaw'].value_counts().reindex(reind).index, (data['DateRaw'].value_counts().reindex(reind)/len(data)).values,color='lightgray')
bars[0].set_color('r'), bars[1].set_color('r'), bars[2].set_color('r'), bars[3].set_color('r'), bars[4].set_color('r')
bars[5].set_color('r'), bars[6].set_color('r'), bars[7].set_color('r'), bars[8].set_color('r'), bars[9].set_color('r')
plt.xticks(rotation=45)
plt.title('Easter: 1583 to 4099\n'+'Discrete, Values=Easter, Probabilities=Probability')
plt.xlabel('Date')
plt.ylabel('Probability')
plt.show()
```

```python
plt.figure(figsize=[12, 8], facecolor='limegreen')
bars = plt.bar(data.dropna()['Years'].astype(int).astype(str).values, data.dropna()['Occurences'], color='deeppink')
plt.xticks(rotation=45)
plt.title('Years Until the First Recurrence of Easter Dates')
plt.xlabel('Years')
plt.ylabel('Occurence')
plt.show()
```
