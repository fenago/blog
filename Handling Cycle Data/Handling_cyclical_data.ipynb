```python
import pandas as pd

df = pd.read_csv('energy_consumption_levels.csv')
df.head()
```

```python
df = df[['consumption', 'hour_of_day']]
```

```python
last_week = df.iloc[-168:]
import matplotlib.pyplot as plt

plt.title('Individual hours', size=20)
plt.plot(range(len(last_week)), last_week['hour_of_day'])
```

```python
import numpy as np
last_week['Sin_Hour'] = np.sin(2 * np.pi * last_week['hour_of_day'] / 
                               max(last_week['hour_of_day']))
last_week['Cos_Hour'] = np.cos(2 * np.pi * last_week['hour_of_day'] /
                               max(last_week['hour_of_day']))
```
